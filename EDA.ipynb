# EDA

```python
import pandas as pd
data = pd.read_csv("../bad_words_data/preprocessing_labeled_data.csv", header=None, names=[1, 2, 3, "label"])
```

```python
data.info() # 데이터 13970개
```

```python
import seaborn as sns
from matplotlib import rc
from matplotlib import rcParams
import matplotlib.pyplot as plt
%matplotlib inline

rc('font', family='AppleGothic', size=11)

plt.rcParams['axes.unicode_minus'] = False
```

# ==== Label 데이터 분포 ====

```python
plt.bar(['욕설', '비욕설'], data['label'].value_counts(), width=0.7)
print(data['label'].value_counts())
# 약 2대1의 비율을 가지고 있음
```

# ==== 데이터 분포 ====

```python
fig = plt.figure(figsize=(27, 7))
ax = plt.subplot(1, 1, 1)

# plt.minorticks_on()
# plt.tick_params(pad=20)
plt.title('가운데 단어 빈도수')
ax.bar(data[2].value_counts().index[:50], data[2].value_counts()[:50], )
for label in ax.xaxis.get_ticklabels() :
    label.set_rotation(45)
    
plt.xlabel('욕설')
plt.ylabel('count')
plt.legend()
plt.show()
# print(data[2].value_counts()[:50])
# 가운데에 욕설이 있는데 욕설의 카운트들이다
```

# ==== 욕설의 형태를 띄지만 욕설이 아닌 데이터의 분포 ====

```python
fig = plt.figure(figsize=(27, 7))
ax = plt.subplot(1, 1, 1)

plt.title('욕설아닌 것의 빈도수')
ax.bar(data[data['label'] == 0][2].value_counts().index[:50], data[data['label'] == 0][2].value_counts()[:50],)
for label in ax.xaxis.get_ticklabels() :
    label.set_rotation(45)
    
plt.xlabel('욕설')
plt.ylabel('count')
plt.legend()
plt.show()
# print(data[2].value_counts()[:50])
# 욕설의 형태를 띄지만 욕설이 아닌 단어들이다.
# 대표적으로 짐승의 새끼를 칭할 때, 프로그램명, 상호명, 신체부위이지만 욕이 아닌 것들, 장애인 대가리 처럼 목적에 따라 다라지는 단어들
```

# ==== 욕설의 형태를 띄면서 욕설인 데이터의 분포 ====

```python
fig = plt.figure(figsize=(27, 7))
ax = plt.subplot(1, 1, 1)

plt.title('욕설아닌 것의 빈도수')
ax.bar(data[data['label'] == 1][2].value_counts().index[:50], data[data['label'] == 1][2].value_counts()[:50],)
for label in ax.xaxis.get_ticklabels():
    label.set_rotation(45)
    
plt.xlabel('욕설')
plt.ylabel('count')
plt.legend()
plt.show()
# print(data[2].value_counts()[:50])
# 욕설의 형태를 띄면서 욕설인 것들은
# 진짜 보기만해도 욕이다.
```
